```python
## Loading the recorded speech file and plotting the signal

import librosa
import matplotlib.pyplot as plt
shreya = r'C:/shreya.wav'
y, sr = librosa.load(shreya)
plt.figure(figsize=(12, 4))
librosa.display.waveshow(y, sr=sr)
plt.title('My First Speech Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.show()
```

```python
## Changing the sampling rate

y, sr = librosa.load(shreya, sr=22000)
plt.figure(figsize=(12, 4))
librosa.display.waveshow(y, sr=sr)
plt.title('Altered Sampling Rate')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.show()
print(f"Default sampling rate is {sr} Hz")
print(f"Sampling Rate after change {sr} Hz")
```

```python
## Observe the length & magnitude range of the signal from the plot. 

print("Finding the Length and the Magnitude of the signal:")
print(f"Signal Length: {len(y)} samples")
print(f"Magnitude Range: {y.min()} to {y.max()}")
```

```python
## Playing the full audio

import sounddevice as sd
print("Playing the full audio")
sd.play(y, samplerate=sr)
sd.wait()
```

```python
## PLaying a part of the signal

from IPython.display import Audio
shreya = r'C:/shreya.wav'
y, sr = librosa.load(shreya)
s = 0
e = 1
print("Playing the first second of the audio")
segment = y[int(s * sr):int(e * sr)]
Audio(data=segment, rate=sr)
```

```python
## segmenting the signal into various components and playing the signal

import IPython.display as ipd
num_segments = 5
segment_duration = len(y) // num_segments
 
for i in range(num_segments):
    start_sample = i * segment_duration
    end_sample = (i + 1) * segment_duration
    segment = y[start_sample:end_sample]
    print(f"Playing the segment {i + 1}:")
    ipd.display(ipd.Audio(segment, rate=sr))
```
